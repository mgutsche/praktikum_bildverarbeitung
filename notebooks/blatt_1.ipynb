## 2.2 Interpolieren in Farbr ̈aumen
_In der Vorlesung haben wir, unter anderem, die Farbräume RGB und HSV kennengelernt._

**a) Was sind die wesentlichen Unterschiede?**

RGB:
- additive Mischung von Rot, Grün und Blau

HSV:
- Darstellung von Farbe als Kombination von Hue (Position auf dem Farbkreis), Saturation (Sättigung) und Value (0=keine Helligkeit, 100=volle Helligkeit)
- Es ist trivial gleich helle Werte im Raum zu finden

**b) Sie sollen in beiden Farbräumen zwischen zwei Farben interpolieren können. Dabei sollen Sie dies möglichst so tun, dass die interpolierten Farben optisch Sinn ergeben (von Gelb nach Rot sollte beispielsweise kein Blau entstehen). Wie schaffen Sie das im RGB Raum und im HSV Raum? Was ist dabei der wesentliche Unterschied. Hinweis: Dies müssen Sie nicht programmieren!**

RGB:
- durchschnitt je Kanal bilden

HSV:
- wesentlicher Unterschied: Farbenkreis fängt nach 360° wieder von Vorne an
    - Also kann ein hue x als x oder x+360 repräsentiert werden
- h3 = Min((h1+h2)/2, (h1+h2+360)/2)
- Für Saturation und Value einfach durchschnitt
