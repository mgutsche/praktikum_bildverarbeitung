## 1.1 Anzeigen eines Bildes in Python
Lade dir ein beliebiges Bild (wenn möglich im JPG Format) aus dem Internet herunter und öffne es in Python. Nutze das Paket Numpy, um aus dem Bild ein Numpy Array zu machen und plotte dieses Array mit dem Paket MatPlotLib.

```python
with Image.open("../images/Mandrill.jpg") as img:
    array = np.asarray(img)
    plt.imshow(array)
```

# 1.2 Tilling
Erweitere nun den Code, in dem du eine Funktion hinzufügst, die das Bild mehrmals vertikal und horizontal kopiert. Hier ein Beispiel:

Wir haben das Bild X gegeben. Die Funktion bekommt als Input die Anzahl der Vertikalen und Horizontalen Kopien, in diesem Fall
Vertikal: 3
Horizontal: 4

Das Ergebnis sollte wie folgt aussehen:
XXXX
XXXX
XXXX

Plotte das Endergebnis mit Matplotlib als ein Bild.

```python
v = 3
h = 4

with Image.open("../images/Mandrill.jpg") as img:
    array = np.asarray(img)
    tiled_array = np.tile(array, (v,h,1))
    plt.imshow(tiled_array)
```

# Cropping
Erweitere dein Programm um eine Funktion, die das Bild nun ausschneidet. Hierbei sollte man sowohl Start als auch Endposition in vertikaler und horizontaler Position angeben können. Plotte das Ergebnis (den ausgeschnittenen Teil) mit Matplotlib. In Zukunft, ist es nicht anders gefordert, solltest du immer das Ergebnis mit Matplotlib darstellen

```python
start = (100, 100)
stop = (400, 400)

with Image.open("../images/Mandrill.jpg") as img:
    array = np.asarray(img)
    tiled_array = array[start[0]:stop[0], start[1]:stop[1]]
    plt.imshow(tiled_array)
```
